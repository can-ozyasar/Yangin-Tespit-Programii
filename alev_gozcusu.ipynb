```python
# İnternet üzerinden veri çekmek için requests kütüphanesini ekliyoruz
import requests

# NASA'nın yangın tespit sistemine erişmek için gerekli bilgileri tanımlıyoruz
# Bu bizim NASA'dan aldığımız özel erişim anahtarımız
API_KEY = "2ece47e141b71c73ecfe8f609c1f8c86"

# Hangi uydudan veri alacağımızı seçiyoruz (VIIRS uydusu)
SOURCE = "VIIRS_SNPP_NRT" 

# Türkiye'nin verilerini almak için ülke kodunu yazıyoruz
COUNTRY = "TUR"

# Son 24 saatin verilerini almak için 1 gün olarak ayarlıyoruz
DAYS = "1" 

# NASA'nın veri sistemine bağlanmak için gerekli web adresini oluşturuyoruz
url = f"https://firms.modaps.eosdis.nasa.gov/api/country/csv/{API_KEY}/{SOURCE}/{COUNTRY}/{DAYS}"

# Kullanıcıya hangi adrese bağlandığımızı gösteriyoruz
print(f"NASA'nın sunucusuna bağlanıyoruz: {url}")

try:
    # NASA'nın sunucusuna bağlanıp veriyi alıyoruz
    response = requests.get(url)

    # Eğer bağlantı başarılı olduysa (200 kodu başarılı bağlantı demek)
    if response.status_code == 200:
        print("\nNASA'nın sunucusuna başarıyla bağlandık. Yangın verilerini indiriyoruz...")
        
        # Aldığımız yangın verilerini bir değişkene kaydediyoruz
        fire_data_text = response.text
        
        # İndirdiğimiz ham veriyi ekranda gösteriyoruz
        print("\n--- İNDİRİLEN HAM VERİ ---")
        print(fire_data_text)
        
        # Verileri bilgisayarımızda bir dosyaya kaydediyoruz
        with open("turkiye_yangin_verisi.csv", "w", encoding="utf-8") as file:
            file.write(fire_data_text)
            
        print("\nYangın verilerini 'turkiye_yangin_verisi.csv' dosyasına kaydettik.")
        
    else:
        # Bağlantı başarısız olursa kullanıcıya hata mesajı gösteriyoruz
        print(f"\nBir sorun oluştu! NASA'nın sunucusuna bağlanamadık. Hata Kodu: {response.status_code}")
        print("NASA'nın sunucusundan gelen mesaj:", response.text)

# Programda başka bir hata olursa (internet bağlantısı kopması gibi) kullanıcıya bildiriyoruz
except Exception as e:
    print(f"Programda bir hata oluştu: {e}")
```
